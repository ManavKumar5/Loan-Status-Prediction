```python
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
data= pd.read_csv("loan_data.csv")
data
```

```python
data.isnull().sum()
```

```python
def fillNullValues(data):
    for column in data.columns:
        if data[column].dtype == "object":
            data[column].fillna(data[column].mode()[0],inplace = True)
        else:
             data[column].fillna(data[column].mean(),inplace = True)

    return "Null values filled"
```

```python
fillNullValues(data)
```

```python
data.isnull().sum()
```

```python
for column in data.columns:
    print("unique values on column{column}\n ")
    print(data[column].unique())
    print("\n------------")
```

```python
#label encoding
```

```python
colname=[]
for x in data.columns:
    if data[x].dtype == "object":
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for x in colname:
    data[x] = le.fit_transform(data[x])
    le_name_mapping = dict(zip(le.classes_,le.transform(le.classes_)))
    print("Feature,x")
    print("mapping",le_name_mapping)
```

```python
data.head()
```

```python
x = data.values[:,1:-1] # Selects all rows (:) and all columns except the last one (0:-1).
y = data.values[:,-1] # Selects all rows (:) and the last column (-1).
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#create a model
model7 = LogisticRegression(class_weight = 'balanced')
```

```python

model7.fit(x_train,y_train)
```

```python
y_pred = model7.predict(x_test)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cfm = confusion_matrix(y_test,y_pred)
print(cfm)

print("classification report:")
print(classification_report(y_test,y_pred))

acc = accuracy_score(y_test,y_pred)
print("Accuracy of the Model:",acc)
```

```python
import pickle
```

```python
with open("model7.pkl","wb")as file:
    pickle.dump(model7,file)
```

```python
def Predict_loan_status(Gender,Married,Dependents,Education,Self_Employed,ApplicantIncome,CoapplicantIncome,LoanAmount,Loan_Amount_Term,Credit_History,Property_Area):
    
    result = model7.predict([[Gender,Married,Dependents,Education,Self_Employed,ApplicantIncome,CoapplicantIncome,LoanAmount,Loan_Amount_Term,Credit_History,Property_Area]])
    return result[0]
```

```python
#over sampling and under sampling for optimizing the model and increase the accuracy#
```

```python
from imblearn.under_sampling import RandomUnderSampler
new_data = RandomUnderSampler(random_state=42)           #reducing more data to balance the model
X, Y= new_data.fit_resample(x, y)
```

```python
model7 = LogisticRegression(random_state=42) # random_state(42)---- random process produces same result/output
model7.fit(X ,Y)                              # training model on resampled data
```

```python
y_pred = model7.predict(x_test)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cfm = confusion_matrix(y_test,y_pred)
print(cfm)

print("classification report:")
print(classification_report(y_test,y_pred))

acc = accuracy_score(y_test,y_pred)
print("Accuracy of the Model:",acc)
```

```python
#store the predicted probabilities
y_pred_prob = model7.predict_proba(x_test)
print(y_pred_prob)
```

```python
y_pred_class=[]
for value in y_pred_prob[:,1]:
    if value > 0.5:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
print(y_pred_class)
        
```

```python
cfm = confusion_matrix(y_test,y_pred_class)
print(cfm)

print("Classification Report")
print(classification_report(y_test,y_pred_class))

accuracy = accuracy_score(y_test,y_pred_class)
print("The accuracy of Model is:",accuracy)
```
